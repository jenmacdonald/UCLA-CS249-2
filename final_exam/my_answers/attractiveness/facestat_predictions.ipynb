```r
d = read.csv("facestat_train.csv", header=TRUE)
test = read.csv("facestat_test.csv", header=TRUE)
```

```r
clean_data = function(d) {
    dataset = d[, -which(colMeans(is.na(d)) > 0.9)]
    dataset_imp = mice(dataset, method = "pmm", m = 5)
    dataset_comp = complete(dataset_imp)
    
    return (dataset_comp)
}
```

```r
dataset_comp = clean_data(d)

dataset_comp$male = ifelse(dataset_comp$male==TRUE,'yes','no')
```

```r
set.seed(604501712)
n = nrow( dataset_comp )
sample.size = 0.75 * n   ###### Use 75% of the data for the training set
training.row.ids = sample( (1:n), sample.size )

my.training.set = dataset_comp[  training.row.ids, ]
my.test.set  = dataset_comp[ -training.row.ids, ]   # set complement of training.set.ids  
```

```r
### Create a control function that will be used across models. We
### create the fold assignments explicitly instead of relying on the
### random number seed being set to identical values.

set.seed(604501712)
ctrl <- trainControl(method = "LGOCV",
                     summaryFunction = twoClassSummary, 
                     classProbs = TRUE,
                     savePredictions = TRUE)
```

```r
set.seed(604501712)
lrFull = train(my.training.set[,-grep("male", colnames(my.training.set))],
             y = my.training.set$male,
             method = "glm",
             metric = "ROC", 
             trControl = ctrl)

confusionMatrix(data = lrFull$pred$pred, reference = lrFull$pred$obs)
predictions = lrFull$pred$pred
predictionsTrans = ifelse(predictions=='yes','TRUE','FALSE')
```

```r
test_dataset_comp = clean_data(test)
```

```r
test_predictions = as.data.frame(predict(lrFull, newdata = test_dataset_comp))
test_predictionsTrans = ifelse(test_predictions=='yes','TRUE','FALSE')

for(i in test_predictionsTrans) {
    cat(i, '\n')
}
```
